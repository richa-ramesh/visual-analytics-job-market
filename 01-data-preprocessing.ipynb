### Schema overview & missing values profile

```python
def profile(df: pd.DataFrame, name: str):
    if df is None: return None
    info = pd.DataFrame({
        'column': df.columns,
        'dtype': [df[c].dtype for c in df.columns],
        'non_null': [df[c].notna().sum() for c in df.columns],
        'nulls': [df[c].isna().sum() for c in df.columns],
        'null_pct': [round(100*df[c].isna().mean(),2) for c in df.columns],
        'nunique': [df[c].nunique(dropna=True) for c in df.columns],
    })
    info['table'] = name
    return info

profiles = []
for name, df in dfs.items():
    p = profile(df, name)
    if p is not None:
        profiles.append(p)
schema_profile = pd.concat(profiles, ignore_index=True)
schema_profile.sort_values(['table','column'])
```

## Dataset Cleaning 

This module provides a structured data-cleaning pipeline for the
*Data Science Job Postings 2025* dataset.

**Functionality:**

- **State Extraction:** Uses regular expressions to extract U.S. state abbreviations from the `headquarter` column and maps them to FIPS codes (`state`, `fips`, `fips_int` columns).
- **Salary Parsing:** Converts salary text ranges (e.g., "$80K–$120K") into numeric midpoints via normalization and numeric extraction (`salary_mid`).
- **Seniority Normalization:** Cleans and standardizes the `seniority_level` column to lowercase values, replacing invalid entries.
- **Skills Parsing:** Transforms stringified skill lists into clean Python lists (`skills_list`) by stripping brackets, quotes, and whitespace.
- **Deduplication:** Removes duplicate job postings using key columns (`job_title`, `company`, `location`, `post_date`, `salary`, `skills`).

```python
from __future__ import annotations
import re
import pandas as pd

# --- Constants ---
STATE_ABBR = r"\b(AL|AK|AZ|AR|CA|CO|CT|DE|FL|GA|HI|ID|IL|IN|IA|KS|KY|LA|ME|MD|MA|MI|MN|MS|MO|MT|NE|NV|NH|NJ|NM|NY|NC|ND|OH|OK|OR|PA|RI|SC|SD|TN|TX|UT|VT|VA|WA|WV|WI|WY|DC)\b"
_ABBR_TO_FIPS = {
    'AL':'01','AK':'02','AZ':'04','AR':'05','CA':'06','CO':'08','CT':'09','DE':'10','DC':'11','FL':'12','GA':'13',
    'HI':'15','ID':'16','IL':'17','IN':'18','IA':'19','KS':'20','KY':'21','LA':'22','ME':'23','MD':'24','MA':'25',
    'MI':'26','MN':'27','MS':'28','MO':'29','MT':'30','NE':'31','NV':'32','NH':'33','NJ':'34','NM':'35','NY':'36',
    'NC':'37','ND':'38','OH':'39','OK':'40','OR':'41','PA':'42','RI':'44','SC':'45','SD':'46','TN':'47','TX':'48',
    'UT':'49','VT':'50','VA':'51','WA':'53','WV':'54','WI':'55','WY':'56'
}

_state_pat = re.compile(STATE_ABBR)


def extract_state(headquarter: pd.Series) -> pd.Series:
    def _one(x):
        s = str(x).upper()
        m = _state_pat.search(s)
        return m.group(1) if m else None
    return headquarter.astype(str).map(_one)


# Salary midpoint parser
def parse_salary(s):
    s = str(s)
    if not s or s.lower() in {"nan", "none"}:
        return None
    s_norm = s.replace("–", "-").lower()  # normalize en dash, lowercase for 'k'
    nums = [float(x.replace(",", "")) for x in re.findall(r"\d[\d,]*\.?\d*", s_norm)]
    if not nums:
        return None
    mid = (nums[0] + nums[1]) / 2 if len(nums) >= 2 else nums[0]
    if "k" in s_norm:
        mid *= 1000.0
    return mid


# skills parser
def parse_skills(s):
    """Parse a simple list-like string into a Python list of skills."""
    if not isinstance(s, str) or not s.strip():
        return []
    return [x.strip(" '\"") for x in s.strip("[]").split(",") if x.strip()]


def clean_jobs_df(df_raw: pd.DataFrame) -> pd.DataFrame:
    df = df_raw.copy()

    # State + FIPS
    df["state"] = extract_state(df.get("headquarter"))
    df["fips"] = df["state"].map(_ABBR_TO_FIPS)  # zero-padded string, e.g., '06'
    df["fips_int"] = pd.to_numeric(df["fips"], errors="coerce").astype("Int64")  # nullable integer

    # Salary
    df["salary_mid"] = df.get("salary", pd.Series([None]*len(df))).apply(parse_salary)

    # Seniority normalization
    df["seniority_level_norm"] = (
        df.get("seniority_level", pd.Series([None]*len(df)))
          .astype(str).str.lower().str.strip().replace({"nan": pd.NA, "": pd.NA})
    )

    # Skills list
    df["skills_list"] = df.get("skills", pd.Series([None]*len(df))).apply(parse_skills)

    # De-duplication
    dedup_keys = ["job_title", "company", "location", "post_date", "salary", "skills"]
    existing_keys = [c for c in dedup_keys if c in df.columns]
    if existing_keys:
        df = df.drop_duplicates(subset=existing_keys)
    return df


if __name__ == "__main__":
    in_path = DATA_DIR /"data_science_job_posts_2025.csv"
    out_path = DATA_DIR /"data_science_job_posts_2025_clean.csv"
    raw = pd.read_csv(in_path, low_memory=False)
    clean = clean_jobs_df(raw)
    clean.to_csv(out_path, index=False)
    print(f"Completed cleaning. new cleaned Dataset created at → {out_path}\n")
```

## 2. Quick summaries, profiling, and cleaning
Below are added code cells that:
- compute column-level profiles (dtypes, non-null counts, null percentages, unique counts)
- show basic numeric summaries and unique value counts for categorical columns
- visualize distributions with Altair (Vega-Lite)
- handle missing values (drop or simple imputation)
- save a cleaned subset for downstream work


```python
# Numeric and categorical summaries
from IPython.display import display, Markdown
def summarize_df(df, name='df', max_unique_cat=30):
    display(Markdown(f'### Summary for `{name}` — {df.shape[0]} rows × {df.shape[1]} cols'))
    display(df.describe(include='all').transpose())
    # separate numeric and categorical
    num_cols = df.select_dtypes(include=['number']).columns.tolist()
    cat_cols = df.select_dtypes(include=['object','category','bool']).columns.tolist()
    display(Markdown(f'**Numeric columns ({len(num_cols)}):** {num_cols}'))
    display(Markdown(f'**Categorical columns ({len(cat_cols)}):** {cat_cols}'))
    # show unique counts for categorical columns (but only up to max_unique_cat unique values)
    for c in cat_cols:
        nunique = df[c].nunique(dropna=True)
        if nunique <= max_unique_cat:
            display(Markdown(f'`{c}` — {nunique} unique: ' + ', '.join(map(str, df[c].dropna().unique()[:max_unique_cat]))))
        else:
            display(Markdown(f'`{c}` — {nunique} unique'))

# run summary for main tables you likely care about
for k in ['postings','companies','salaries']:
    if k in dfs:
        summarize_df(dfs[k], k)
```

```python
# Cleaning: inspect missingness and apply simple strategies
from IPython.display import display, Markdown
def missing_report(df, name='df', show_top=10):
    display(Markdown(f'### Missingness report for `{name}`'))
    miss = pd.DataFrame({
        'column': df.columns,
        'dtype': [df[c].dtype for c in df.columns],
        'non_null': [df[c].notna().sum() for c in df.columns],
        'nulls': [df[c].isna().sum() for c in df.columns],
        'null_pct': [100*df[c].isna().mean() for c in df.columns],
        'nunique': [df[c].nunique(dropna=True) for c in df.columns],
    }).sort_values('null_pct', ascending=False)
    display(miss.head(show_top))
    return miss

# Example: run on postings
if 'postings' in dfs:
    miss_post = missing_report(dfs['postings'], 'postings', show_top=30)

    # strategy: drop columns with >90% missing, drop rows with critical missing data (e.g., job_id)
    cols_to_drop = miss_post[miss_post['null_pct']>90]['column'].tolist()
    print('Columns to drop (>90% missing):', cols_to_drop)
    cleaned = dfs['postings'].copy().drop(columns=cols_to_drop)
    # drop rows without job_id or title
    critical = [c for c in ['job_id','title','company_name'] if c in cleaned.columns]
    cleaned = cleaned.dropna(subset=critical, how='any')
    print('After dropping, shape:', cleaned.shape)
    # simple imputation example: fill missing numeric with median, categorical with 'Unknown'
    for c in cleaned.select_dtypes(include=['number']).columns:
        med = cleaned[c].median()
        cleaned[c] = cleaned[c].fillna(med)
    for c in cleaned.select_dtypes(include=['object','category']).columns:
        cleaned[c] = cleaned[c].fillna('Unknown')
    # save a sampled subset for faster future work
    sample = cleaned.sample(min(20000, len(cleaned)), random_state=42)
    sample.to_csv('postings_clean_sample.csv', index=False)
    print('Saved sample to postings_clean_sample.csv')
else:
    print('No postings table present; skipping cleaning step.')
```
